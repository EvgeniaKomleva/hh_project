```python
import requests

base = {
    'username':'def',
    'password':'def'
}

login = input('Enter login: ')
passwd = input('Enter password: ')

base['username'] = login
base['password'] = passwd

session = requests.Session()
url = 'https://hh.ru/account/login?backurl=%2F'
auth = session.post(url, data=base)
f = open('result1.txt','w+')
f.write(auth.text)
f.close()
```
